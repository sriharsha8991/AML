### BUILD THE MODEL

It requires configuring the layers of the model, then compilinh the model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation = 'relu'),
    tf.keras.layers.Dense(10)
])
```

## Layers 
The basic building block of a neural network is the **layer**.

The first layer in a deep learning network is often used to pre-process the input data so that it can be effectively fed into the rest of the network for processing. The **Flatten layer** is a common layer used for this purpose in image classification problems.

The input data in an image classification problem is typically a 2D array of pixels, with each pixel representing the color of a single point in the image. This array has a width and height, and together they form the image's resolution. For example, in the MNIST dataset, the images are 28x28 pixels, so the input data is a 2D array of size 28x28.

However, for the purposes of a deep learning network, **this 2D array must be transformed into a 1D array**. This is where the Flatten layer comes in. The Flatten layer takes the 2D array of pixels and "unstacks" them, so to speak, transforming the data into a 1D array. This **1D array then represents the entire image as a single line of pixel values**.

For the MNIST dataset, the images are 28x28 pixels, so the Flatten layer would take the 2D array of 28x28 pixels and transform it into a 1D array of 784 pixels (28 x 28 = 784). This 1D array is then fed into the next layer in the network for processing.

It's important to note that the Flatten layer has no parameters to learn. It simply reformats the input data into a structure that can be used by the rest of the network. *This layer has no impact on the model's ability to learn and make predictions, it simply pre-processes the input data so that it can be effectively processed by the rest of the network*.

)

In a deep learning network for image classification, the Flatten layer is often followed by one or more fully connected (also known as dense) layers. In this case, the network consists of two Dense layers.

A **Dense layer** is a type of neural network layer that is fully connected, meaning that every node in the layer is connected to every node in the previous layer. Each node in the Dense layer represents a neuron that receives inputs from the previous layer, processes those inputs using an activation function, and passes the results to the next layer

### Compiling the model

1. Loss function â€”This measures how accurate the model is during training. You want to minimize this function to "steer" the model in the right direction.

2. Optimizer â€”This is how the model is updated based on the data it sees and its loss function.

3. Metrics â€”Used to monitor the training and testing steps. The following example uses accuracy, the fraction of the images that are correctly classified

```python
model.compile(optimizer = 'adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

### Feed the model with training data

```python
model.fit(train_images,train_labels,epochs=15)
```

an **epoch** is one full iteration over the entire training dataset. During each epoch, the model is presented with each sample in the training dataset, and its parameters are updated to minimize the error between its predictions and the actual target values.

```python
test_loss,test_acc = model.evaluate(test_images,test_labels,verbose=2)
print("\nTest accuracy:",test_acc)
```

#### Make Predictions

```python
probability_model = tf.keras.Sequential([
    tf.keras.layers.Softmax()
])
```

This above code snippet is TensorFlow Keras model with a single layer,***a softmax activation layer.*** 

The softmax function is commonly used in the output layer of a neural network to produce a probability distribution over multiple classes.

The input to this layer is expected to be an array of real numbers and the output of the layer is a probability distribution over multiple classes, where the sum of all probabilities is equal to 1. 

The purpose of this model is likely to be used for a __multi-class classification problem__, where the goal is to predict the most likely class given an input.

```python
predictions = probability_model.predict(test_images)
```

The model has predicted the label for each image in the testing set
